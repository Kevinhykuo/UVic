```python
import numpy as np 
import matplotlib as mp
import matplotlib.pyplot as plt

from numpy import sin, pi
from scipy.optimize import curve_fit 
from textwrap import wrap

%matplotlib inline
%config InlineBackend.figure_format = 'pdf'
```

## Question 1

```python
P = np.array([[-2, 0], [-1, -1], [0, 0], [1, 1], [2,0]])
x_space = np.linspace(-2, 2, 100)
```

```python
def f1(x, a, b, c, d, e):
    f = a*x**4 + b*x**3 + c*x**2 + d*x + e
    return f
```

```python
popt1, pcov1 = curve_fit(f1, P[:,0], P[:,1])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

ax.scatter(P[:,0], P[:,1], c='black', label="Data points", zorder=5)
ax.plot(x_space, sin(2*x_space / pi), label=r"$y=\sin \left(\frac{2 x}{\pi}\right)$")
ax.plot(x_space, f1(x_space, *popt1), '-', 
         label=r'Polynomial fit ($y=ax^4+bx^3+cx^2+dx+e$): $a$=%5.3f, $b$=%5.3f, $c$=%5.3f, $d$=%5.3f, $e$=%5.3f' % tuple(popt1))

ax.set_ylabel(r"$y$", rotation=0, fontsize=14)
ax.set_xlabel(r"$x$", fontsize=14)

plt.ylim(-1.75, 1.25)
plt.legend()
plt.show()
```

## Question 3

```python
paths = 100
time = 50 
RW = np.random.randint(0, 2, size=(paths, time)) * 2 - 1
RW_pos = np.cumsum(RW, axis=1)

t = np.arange(0, np.shape(RW_pos)[0], 1)
```

```python
fig, ax3 = plt.subplots(1, 1, figsize=(10, 8))

for i in range(0, time):
    plt.plot(t, RW_pos[:,i])
    i += 1
    
plt.show()
```
